## Data Preprocessing
#### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import warnings
warnings.simplefilter(action = 'ignore', category = Warning)
```

#### Importing Dataset

```python
dataset = pd.read_csv('heart.csv')
```

```python
dataset
```

```python
dataset = dataset.drop_duplicates()
```

#### Create Feature Matrix and Dependent Variable Vector

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

#### Replace Missing Data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x)
x = imputer.transform(x)
```

#### Spliting the Dataset into Training Dataset and Testing Dataset

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
Xtest
```

```python
Ytrain
```

```python
Ytest
```

#### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
Xtrain = ss.fit_transform(Xtrain)
Xtest = ss.fit_transform(Xtest)
```

## Building Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
LoR = LogisticRegression(random_state=0)
LoR.fit(Xtrain, Ytrain)
```

```python
Yestimated = LoR.predict(Xtest)
Yestimated
```

```python
# Performance Metric
from sklearn.metrics import confusion_matrix, accuracy_score
print("Confusion Matrix:\n", confusion_matrix(Ytest, Yestimated))
print("Accuracy Score:", accuracy_score(Ytest, Yestimated))
```

## Building KNN Classification Model

```python
from sklearn.neighbors import KNeighborsClassifier
KC = KNeighborsClassifier(n_neighbors=10, weights='uniform', p=2)
KC.fit(Xtrain, Ytrain)
```

```python
Yest = KC.predict(Xtest)
```

```python
#Performance Metric
from sklearn.metrics import confusion_matrix,accuracy_score
print("Confusion Matrix:\n", confusion_matrix(Ytest, Yest))
print("Accuracy Score:", accuracy_score(Ytest, Yest))
```

```python
errorRate=[]
for i in range(1,30):
 KC = KNeighborsClassifier(n_neighbors = i)
 KC.fit(Xtrain, Ytrain)
 Ypred_i = KC.predict(Xtest)
 errorRate.append(np.mean(Ypred_i != Ytest))
```

```python
graph2=plt.plot(range(1,30), errorRate, marker='o', markerfacecolor='black', markersize=5)
plt.xlabel('K-Value')
plt.ylabel('Error-Rate')
```
